```python
%matplotlib inline
import numpy as np
import cv2
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [9, 7]
```

We used a blue filter on our camera so the blue channel is populated with NIR, so

$$ ndvi = \frac{nir - blue}{nir + blue} $$

```python
img = cv2.imread("succ_raw.jpg")
r_img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(r_img)
nir, g, b = r_img[:,:,0], r_img[:,:,1], r_img[:,:,2]
ndvi = (nir - b)/(nir+b)
plt.imshow(ndvi, cmap="gray")
plt.savefig("first_ndvi_lol.png")
```
