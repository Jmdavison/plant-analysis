```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math

np.seterr(divide='ignore', invalid='ignore')
plt.rcParams['figure.figsize'] = [10, 10]
```

```python
img = cv2.imread("data/sat/sat_0.png")
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(img)
plt.show()
```

```python
blur = cv2.GaussianBlur(img,(5,5),0)
# plt.imshow(blur)
# plt.show()
color = ('r','g','b')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
plt.show()
```

```python
r,g,b = img[::,::,0].astype('float'), img[::,::,1].astype('float'), img[::,::,2].astype('float')

ndvi = (b - r)/(b + r)
plt.imshow(ndvi, cmap="jet")
ndvi[np.isnan(ndvi)] = 0
ndvi *= (ndvi > 0)
ndvi_scaled = (ndvi * 255).astype('uint8')
print(ndvi_scaled.max())
ax = plt.gca()
ax.imshow(ndvi_scaled,cmap="jet")
plt.colorbar()
```

```python
# load fasties color map
cmap = np.loadtxt('colormap.txt', dtype=np.uint8)

print(ndvi_scaled)
print(cmap[ndvi_scaled.max(), 1:4])
cmapped_ndvi_unstretched = cmap[ndvi_scaled, 1:4]
cv2.imwrite('fastie-weird.png', cmapped_ndvi_unstretched)
plt.imshow(cmapped_ndvi_unstretched)
plt.show()
```

```python
edged = cv2.Canny(blur, 10,170,True, 3)
plt.imshow(edged)
plt.show()
```

```python
kern = cv2.getStructuringElement(cv2.MORPH_RECT, (23,23))
dilated = cv2.morphologyEx(edged, cv2.MORPH_DILATE, kern)
kern = cv2.getStructuringElement(cv2.MORPH_RECT, (18,18))
eroded = cv2.morphologyEx(dilated, cv2.MORPH_ERODE, kern)
plt.imshow(eroded)
```

```python
cnt_img, contours, dtype = cv2.findContours(eroded.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
areas = [cv2.contourArea(c) for c in contours]
biggest_ind = np.argmax(areas)
circ = cv2.minEnclosingCircle(contours[biggest_ind])
roi = cv2.circle(
    img.copy(), 
    (int(circ[0][0]), int(circ[0][1])), 
    int(circ[1]),
    (255,255,0),
    10
)
plt.imshow(roi)
```

```python
#create a mask
mask = np.zeros(img.shape[:2], np.uint8)
cv2.circle(
    mask, 
    (int(circ[0][0]), int(circ[0][1])), 
    int(circ[1]),
    1,
    -1,
    4
)
cut = img*mask[:,:,np.newaxis]
plt.imshow(cut)
```

```python
r,g,b = cut[:,:,0].astype('float'), cut[:,:,1].astype('float'), cut[::,::,2].astype('float')

print(r.max(),g.max(),b.max())

ndvi_cut = (b - r)/(b + r)
ndvi_cut[np.isnan(ndvi_cut)] = 0
ndvi *= (ndvi > 0)
print(ndvi_cut)
ndvi_cut_scaled = (ndvi_cut * 255).astype(np.uint8)
plt.imshow(ndvi_cut_scaled)
plt.show()
```

```python
ndvi_renorm = (ndvi_cut - ndvi_cut.min())/(ndvi_cut.max() - ndvi_cut.min())
# rescale and show
ndvi_view = (ndvi_renorm * 255).astype(np.uint8)
cv2.imwrite('test_newfilt.png', ndvi_view)
plt.imshow(ndvi_view)
plt.show()
```

```python
# load fasties color map
cmap = np.loadtxt('colormap.txt', dtype=np.uint8)
cmapped_ndvi = cmap[ndvi_view, 1:4]
cv2.imwrite('fastie-newfilt.png', cv2.cvtColor(cmapped_ndvi,cv2.COLOR_BGR2RGB))
print(ndvi_view)
plt.imshow(cmapped_ndvi)
plt.show()
```
